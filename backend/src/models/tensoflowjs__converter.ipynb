```python
# Step 2: Import necessary libraries
import tensorflow as tf
import tensorflowjs as tfjs
```

```python
# Step 3: Load the trained Keras model
model = tf.keras.models.load_model("/content/trained_model.keras")

# Optional: Check model summary
model.summary()

# Step 4: Convert and save the model to TensorFlow.js format
tfjs_target_dir = "tfjs_model1"

# Save the Keras model as a TensorFlow SavedModel first
saved_model_dir = "saved_model_for_tfjs"
tf.saved_model.save(model, saved_model_dir)

# Convert the TensorFlow SavedModel to TensorFlow.js format
tfjs.converters.convert_tf_saved_model(saved_model_dir, tfjs_target_dir)


print(f"Model successfully converted and saved to: {tfjs_target_dir}")
```
